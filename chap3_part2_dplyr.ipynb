```r
# Exercise 

flights_sml <- select(flights,
                        year:day,
                      arr_time,
                      dep_time,
                      air_time
                     )
# mutate(flights_sml,
#       my_air_time = arr_time - dep_time
#       ) # both will not match

myfunc <- function(x){
     x %/% 100 * 60 + x %% 100 
}
mutate(flights_sml,
        arr_time_new = myfunc(air_time),
        dep_time_new = myfunc(dep_time),
        my_air_time = dep_time_new - arr_time_new
       )
```

```r
# Exercise 

flights_sml <- select(flights,
                      arr_time,
                      dep_time,
                      air_time,
                      sched_arr_time,
                      sched_dep_time
                     )
# mutate(flights_sml,
#       my_air_time = arr_time - dep_time
#       ) # both will not match

myfunc <- function(x){
     x %/% 100 * 60 + x %% 100 
}
mutate(flights_sml,
        arr_time_new = myfunc(air_time),
        sched_arr_time = myfunc(sched_arr_time),
        sched_dep_time = myfunc(sched_dep_time),
        dep_time_new = myfunc(dep_time),
        my_air_time = dep_time_new - arr_time_new
       )
```

```r
flights %>% 
  mutate(dep_time = (dep_time %/% 100) * 60 + (dep_time %% 100),
         sched_dep_time = (sched_dep_time %/% 100) * 60 + (sched_dep_time %% 100),
         arr_time = (arr_time %/% 100) * 60 + (arr_time %% 100),
         sched_arr_time = (sched_arr_time %/% 100) * 60 + (sched_arr_time %% 100)) %>%
  transmute((arr_time - dep_time) %% (60*24) - air_time)
```

```r
by_dest <- group_by(flights, dest)
# str(by_dest)

delay <- summarize(by_dest,
count = n(),
dist = mean(distance, na.rm = TRUE),
delay = mean(arr_delay, na.rm = TRUE)
)

delay <- filter(delay, count > 20, dest != "HNL")
# str(delay)
# fix(delay)

ggplot(data = delay, mapping = aes(x = dist, y = delay)) +
geom_point(aes(size = count), alpha = 1/3) +
geom_smooth(se = F)
```
