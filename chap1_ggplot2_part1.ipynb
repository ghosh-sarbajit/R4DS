```r
table((ggplot2::mpg)$class)
table((ggplot2::mpg)$cyl)
summary((ggplot2::mpg)$displ)
summary((ggplot2::mpg)$hwy)
table((ggplot2::mpg)$drv)

ggplot(data = mpg) +
geom_point(mapping = aes(x = displ, y = hwy)) +
facet_wrap(~ class, nrow = 2)

ggplot(data = mpg) +
geom_point(mapping = aes(x = displ, y = hwy)) +
facet_grid(drv ~ cyl)
```

```r
# geometry

ggplot(data = mpg) +
geom_point(mapping = aes(x = displ, y = hwy))

ggplot(data = mpg) +
geom_smooth(mapping = aes(x = displ, y = hwy))

ggplot(data = mpg) +
geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv))


# contains duplication
ggplot(data = mpg) +
geom_smooth(
    mapping = aes(x = displ, y = hwy, linetype = drv, color = drv),
    show.legend = TRUE
) +
geom_point(mapping = aes(x = displ, y = hwy, color = drv))
```

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy, color = drv)) +
geom_point() +
geom_smooth()

ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) +
geom_point(mapping = aes(color = class)) +
geom_smooth()
```

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) +
geom_point(mapping = aes(color = class)) +
geom_smooth(
data = filter(mpg, class == "subcompact"),
se = TRUE # displays confidence interval around smoothing
)
```

```r
# exercise page 20

# box plot
ggplot(data = mpg) +
geom_boxplot(mapping = aes( hwy, model)) 

ggplot(data = mpg, aes(hwy)) +
geom_boxplot() +
facet_grid(drv ~ cyl)

ggplot(data = mpg, aes(hwy)) +
geom_boxplot() +
facet_grid(drv ~ .)

# line chart
ggplot(data = mpg, aes(x = hwy, y = year)) +
geom_line() +
geom_smooth(
    data = filter(mpg, manufacturer == "toyota")
)

# histogram

ggplot(mpg, aes(hwy)) +
  geom_histogram(binwidth = 1)

# area chart

ggplot(mpg, aes(x=hwy, y=displ)) +
  geom_area( fill="#69b3a2", alpha=0.4)
```
